```python
dataset = get_data("diamond")
```

```python
#price is the target variable, the rest are independent variables.
```

```python
dataset.describe()
```

```python
data_train = dataset.sample(frac=0.9, random_state=786)
# Select a random sample of 90% of the data for training and set random state to any number
```

```python
data_train.describe()
```

```python
data_train.head()
```

```python
data_test = dataset.drop(data_train.index)
#drop all samples taken for training 
```

```python
data_test.shape
```

```python
data_train.reset_index(drop = True, inplace = True) #training dataset in proper order
```

```python
data_train.head()
```

```python
data_test.reset_index(drop = True, inplace = True) #testing dataset in proper order
```

```python
data_test.head()
```

```python
from pycaret.regression import *
```

```python
reg = setup(data= data_train, target = "Price", session_id = 123)
```

```python
best = compare_models(exclude = ["ransac"])
```

```python
ada = create_model("ada")
```

```python
tuned_ada = tune_model(ada) #fine tuning the model after creating it using tune_model 
```

```python
plot_model(tuned_ada) # we can see that r^2 has increased after fine tuning the model
```

```python
plot_model(tuned_ada, plot = "error") #plotting the error plot for this model 
```

```python
plot_model(tuned_ada, plot= "feature") # shows which features are most important 
```

```python
predict_model(tuned_ada) 
```

```python
test_pred = predict_model(tuned_ada, data = data_test)
```

```python
test_pred.head()
```

```python
save_model(tuned_ada, "Regression Model") #saves the model and you can name it anything
```

```python
saved_final_model = load_model("Regression Model") #loading the model that we saved
```

```python
new_prediction = predict_model(saved_final_model, data = data_test)
```
